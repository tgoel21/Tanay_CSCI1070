```python
titanic_df["mr_mrs"]=np.where(titanic_df["Name"].str.contains("Mrs."),"Mrs",
                              np.where(titanic_df["Name"].str.contains("Ms."),"Ms","Mr"))
# this is not accurate to assume everyone who doesnt have Mrs. or Ms. in
# their name is a Mr
```

# Histograms

```python
# histograms show us frequency of whatever column we pass in.
# in other words: how many records have that value
# pandas plot
embarked_count= titanic_df["Embarked"].value_counts()

embarked_count.plot(kind="bar")
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Histogram of Titanic Ports of Embarkation')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.grid()
```
