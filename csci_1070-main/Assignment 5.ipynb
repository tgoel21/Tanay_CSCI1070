```python
# (6 pts) Perform pre-processing on the dataset. Clearly use markdown to indicate each pre-processing step you are using. Note: you should use at least three different techniques. Depending on your dataset you may need to create new columns from the existent data you have.
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
df = pd.read_csv('loans_full_schema.csv')
imputer = SimpleImputer(strategy='median')
df[['annual_income',
    'debt_to_income',
    'annual_income_joint',
    'debt_to_income_joint',
    'total_credit_limit',
    'total_credit_utilized',
    'total_collection_amount_ever',
    'total_debit_limit',
    'loan_amount',
    'term',
    'interest_rate',
    'installment',
    'balance',
    'paid_total',
    'paid_principal',
    'paid_interest',
    'paid_late_fees']] = imputer.fit_transform(df[['annual_income',
                                                   'debt_to_income',
                                                   'annual_income_joint',
                                                   'debt_to_income_joint',
                                                   'total_credit_limit',
                                                   'total_credit_utilized',
                                                   'total_collection_amount_ever',
                                                   'total_debit_limit',
                                                   'loan_amount',
                                                   'term',
                                                   'interest_rate',
                                                   'installment',
                                                   'balance',
                                                   'paid_total',
                                                   'paid_principal',
                                                   'paid_interest',
                                                   'paid_late_fees']])

imputer = SimpleImputer(strategy='most_frequent')
df[['emp_title',
    'state',
    'homeownership',
    'verified_income',
    'verification_income_joint',
    'loan_purpose',
    'application_type',
    'grade',
    'sub_grade',
    'issue_month',
    'loan_status',
    'initial_listing_status',
    'disbursement_method']] = imputer.fit_transform(df[['emp_title',
                                                        'state',
                                                        'homeownership',
                                                        'verified_income',
                                                        'verification_income_joint',
                                                        'loan_purpose',
                                                        'application_type',
                                                        'grade',
                                                        'sub_grade',
                                                        'issue_month',
                                                        'loan_status',
                                                        'initial_listing_status',
                                                        'disbursement_method']])
le = LabelEncoder()
categorical_cols = ['emp_title',
                    'state',
                    'homeownership',
                    'verified_income',
                    'verification_income_joint',
                    'loan_purpose',
                    'application_type',
                    'grade',
                    'sub_grade',
                    'issue_month',
                    'loan_status',
                    'initial_listing_status',
                    'disbursement_method']
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
scaler = StandardScaler()
df[['annual_income',
    'debt_to_income',
    'annual_income_joint',
    'debt_to_income_joint',
    'total_credit_limit',
    'total_credit_utilized',
    'total_collection_amount_ever',
    'total_debit_limit',
    'loan_amount',
    'term',
    'interest_rate',
    'installment',
    'balance',
    'paid_total',
    'paid_principal',
    'paid_interest',
    'paid_late_fees']] = scaler.fit_transform(df[['annual_income',
                                                  'debt_to_income',
                                                  'annual_income_joint',
                                                  'debt_to_income_joint',
                                                  'total_credit_limit',
                                                  'total_credit_utilized',
                                                  'total_collection_amount_ever',
                                                  'total_debit_limit',
                                                  'loan_amount',
                                                  'term',
                                                  'interest_rate',
                                                  'installment',
                                                  'balance',
                                                  'paid_total',
                                                  'paid_principal',
                                                  'paid_interest',
                                                  'paid_late_fees']])
df.head()
```

```python
# (3 pts) If using the Lending Club dataset create subplot histograms to show if loan_amounts differ by region. If using the Titanic dataset, create subplot histograms to show how fares differ by age_bin. 
# Summarize your findings in under 200 words.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('loans_full_schema.csv')
region_map = lambda state: {
    'AK': 'Pacific', 'WA': 'Pacific', 'OR': 'Pacific', 'CA': 'Pacific', 'NV': 'Pacific', 'AZ': 'Pacific', 'HI': 'Pacific',
    'MT': 'Plains', 'WY': 'Plains', 'ND': 'Plains', 'SD': 'Plains', 'MN': 'Plains', 'IA': 'Plains', 'KS': 'Plains', 'MO': 'Plains', 'NE': 'Plains', 'CO': 'Plains',
    'IL': 'Midwest', 'IN': 'Midwest', 'MI': 'Midwest', 'OH': 'Midwest', 'WI': 'Midwest',
    'ME': 'Northeast', 'NH': 'Northeast', 'VT': 'Northeast', 'MA': 'Northeast', 'RI': 'Northeast', 'CT': 'Northeast', 'NY': 'Northeast', 'NJ': 'Northeast', 'PA': 'Northeast',
    'TX': 'South', 'OK': 'South', 'AR': 'South', 'LA': 'South', 'MS': 'South', 'AL': 'South', 'GA': 'South', 'FL': 'South', 'SC': 'South', 'NC': 'South', 'VA': 'South', 'WV': 'South', 'KY': 'South', 'TN': 'South'
}.get(state, 'Unknown')
df['region'] = df['state'].apply(region_map)
fig, axs = plt.subplots(1, 5, figsize=(20, 6))
regions = ['Pacific', 'Plains', 'Midwest', 'Northeast', 'South']
for x, region in enumerate(regions):
    region_df = df[df['region'] == region]
    sns.histplot(region_df['loan_amount'], ax=axs[x], bins=50, kde=True)
    axs[x].set_title(region)
    axs[x].set_xlabel('Loan Amount')
    axs[x].set_ylabel('Frequency')
fig.tight_layout()
plt.show()

# The histograms show the loan amounts and the frequency of these loans in different regions and determine a line of best fit for the data allowing us to determine trends in the data.
```
