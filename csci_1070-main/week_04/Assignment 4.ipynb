```python
import numpy as np
import matplotlib.pyplot as plt

def plot_major(studentsurveydf, title='Major', xlabel='Major', ylabel='Count', rotation=90):
    major_counts = studentsurveydf['What is your major?'].value_counts()
    major_index = major_counts.index
    x = np.array(major_index)
    y = np.array(major_counts)
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.xticks(rotation=rotation)
    plt.show()

def plot_num_pets(studentsurveydf, title='Number of Pets', xlabel='Number of Pets', ylabel='Count', bins=10):
    num_pets = studentsurveydf['How many pets and of which kind do you have?']
    x = np.array(range(1, bins+1))
    y = np.array([sum(num_pets == i) for i in range(1, bins+1)])
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

def plot_age_vs_num_pets(studentsurveydf, title='Age vs Number of Pets', xlabel='Age', ylabel='Number of Pets'):
    age = studentsurveydf['How old are you?']
    num_pets = studentsurveydf['How many pets and of which kind do you have?']
    x = np.array(age)
    y = np.array(num_pets)
    plt.scatter(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

def plot_favorite_food(studentsurveydf, title='Favorite Food', xlabel='Food', ylabel='Count'):
    favorite_food = studentsurveydf['What is your favorite type of food?'].value_counts()
    labels = favorite_food.index
    x = np.array(labels)
    y = np.array(favorite_food)
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

def plot_country_birth(studentsurveydf, title='Country of Birth', xlabel='Country', ylabel='Count'):
    country_birth = studentsurveydf['What state/country were you born in?'].value_counts()
    labels = country_birth.index
    x = np.array(labels)
    y = np.array(country_birth)
    plt.bar(x, y)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

plot_major(studentsurveydf)
plot_num_pets(studentsurveydf)
plot_age_vs_num_pets(studentsurveydf)
plot_favorite_food(studentsurveydf)
plot_country_birth(studentsurveydf)
```
